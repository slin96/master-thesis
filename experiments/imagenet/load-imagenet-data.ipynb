```python
import torchvision
import numpy as np
from matplotlib.pyplot import imshow
```

```python
%matplotlib inline
```

```python
root_path = '/hpi/fs00/share/fg/rabl/strassenburg/datasets/imgnet'
imagenet_data = torchvision.datasets.ImageNet(root_path, split='val')
```

```python
def show_image(element):
    img, cls = element
    image_class = imagenet_data.classes[cls]
    print('image class: {}'.format(image_class))

    imshow(np.asarray(img))
```

```python
show_image(imagenet_data[0])
```

```python
show_image(imagenet_data[4200])
```
